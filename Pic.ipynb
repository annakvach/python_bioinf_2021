```python
from PIL import Image, ImageFilter
import numpy as np
from IPython.display import display
import ipyplot
import matplotlib.pyplot as plt
```

https://pillow.readthedocs.io/en/stable/ - ссылка на документацию pillow

https://github.com/karolzak/ipyplot - ссылка на документацию ipyplot

```python
path = '/mnt/c/my_mnt_c/Data/i.jpg'
```

```python
im = Image.open(path)
```

```python
print("Размер изображения: ")  
print(im.format, im.size, im.mode)
```

```python
blur = im.filter(ImageFilter.BLUR)
contour = im.filter(ImageFilter.CONTOUR)
detail = im.filter(ImageFilter.DETAIL)
edge_enhance = im.filter(ImageFilter.EDGE_ENHANCE)
edge_enhance_more = im.filter(ImageFilter.EDGE_ENHANCE_MORE)
emboss = im.filter(ImageFilter.EMBOSS)
find_edges = im.filter(ImageFilter.FIND_EDGES)
sharpen = im.filter(ImageFilter.SHARPEN)
smooth = im.filter(ImageFilter.SMOOTH)
smooth_more = im.filter(ImageFilter.SMOOTH_MORE)
```

```python
images_array = [im, 
                blur, contour, detail,
                edge_enhance, edge_enhance_more, emboss, 
                find_edges,sharpen,smooth,
                smooth_more]
```

```python
labels = ["original", 
                "blur", "contour", "detail",
                "edge_enhance", "edge_enhance_more", "emboss", 
                "find_edges","sharpen","smooth",
                "smooth_more"]
```

```python
ipyplot.plot_images(images_array, labels, max_images=20, img_width=400)
```
